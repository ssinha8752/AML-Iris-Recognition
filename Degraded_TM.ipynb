```python
# COMPLETE IRIS RECOGNITION PIPELINE WITH ENHANCED FEATURE EXTRACTION
import pandas as pd
import numpy as np
import cv2
import os
import time
from skimage.feature import hog, local_binary_pattern
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
import matplotlib.pyplot as plt
import joblib

# 1. Load and prepare dataset
print("Loading dataset...")
try:
    df = pd.read_csv('/content/drive/MyDrive/AML-PROJECT/iris_degrade.csv')
    print("Data loaded successfully. Columns:", df.columns.tolist())
except Exception as e:
    print("Error loading CSV:", e)
    raise

# 2. Extract IDs and filter
print("\nPreprocessing data...")
df['ID'] = df['Label'].str.extract(r'(\d+)', expand=False)
valid_ids = df['ID'].value_counts()[df['ID'].value_counts() >= 10].head(100).index
df = df[df['ID'].isin(valid_ids)]
print(f"Filtered to {len(df)} samples from {len(valid_ids)} IDs")

# 3. Smart path resolution with degraded directory
print("\nResolving image paths...")
BASE_PATH = '/content/drive/MyDrive/AML-PROJECT/'
DEGRADED_DIR = 'degraded/'

def resolve_image_paths(df):
    """Handle multiple possible path configurations"""
    if 'Path' in df.columns:
        print("Using existing 'Path' column")
        return df['Path']

    if 'Label' in df.columns:
        print("Constructing paths from Label column")
        paths = []
        for label in df['Label']:
            possible_paths = [
                os.path.join(BASE_PATH, DEGRADED_DIR, f"{label}.jpg"),
                os.path.join(BASE_PATH, DEGRADED_DIR, label),
                os.path.join(BASE_PATH, f"{label}.jpg"),
                os.path.join(BASE_PATH, label)
            ]

            for path in possible_paths:
                if os.path.exists(path):
                    paths.append(path)
                    break
            else:
                paths.append(None)
                print(f"Warning: No valid path found for label {label}")
        return pd.Series(paths)

    raise ValueError("Could not determine image paths - no valid columns found")

df['resolved_path'] = resolve_image_paths(df)
valid_images = df[df['resolved_path'].notna()]
print(f"Found {len(valid_images)} valid image paths")

# 4. Enhanced image preprocessing
def preprocess_image(img, target_size=(120, 160)):
    """Apply histogram equalization and blurring"""
    # Convert to grayscale if needed
    if len(img.shape) > 2:
        img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Resize and apply preprocessing
    img = cv2.resize(img, target_size)
    img = cv2.equalizeHist(img)  # Histogram equalization
    img = cv2.GaussianBlur(img, (5, 5), 0)  # Gaussian blur
    return img

# 5. Feature extraction pipeline
def extract_enhanced_features(images):
    """Extract HOG + LBP features with PCA"""
    features = []
    error_count = 0

    for i, img in enumerate(images):
        try:
            # Preprocess image
            img = preprocess_image(img)

            # HOG Features
            hog_features = hog(
                img, orientations=8, pixels_per_cell=(16, 16),
                cells_per_block=(1, 1), visualize=False
            )

            # LBP Features
            lbp = local_binary_pattern(img, P=8, R=1, method='uniform')
            lbp_hist = np.histogram(lbp, bins=256, range=(0, 256))[0]

            # Combine features
            combined = np.concatenate([hog_features, lbp_hist])
            features.append(combined)
        except Exception as e:
            error_count += 1
            print(f"Error processing image {i}: {str(e)}")
            features.append(np.zeros(128 + 256))  # HOG + LBP dimensions

    if error_count > 0:
        print(f"\nEncountered {error_count} feature extraction errors")

    return np.array(features)

# 6. Load and preprocess images
print("\nLoading and preprocessing images...")
def load_images(df, target_size=(120, 160)):
    images, labels = [], []
    start_time = time.time()
    load_count = 0

    for i, (path, label) in enumerate(zip(df['resolved_path'], df['ID'])):
        try:
            if not os.path.exists(path):
                continue

            img = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
            if img is None:
                continue

            img = preprocess_image(img, target_size)
            images.append(img)
            labels.append(label)
            load_count += 1

            if (i + 1) % 100 == 0:
                elapsed = time.time() - start_time
                print(f"Processed {i + 1} images | Loaded {load_count} | Elapsed: {elapsed:.1f}s")
        except Exception as e:
            print(f"[Error] Processing {path}: {str(e)}")
            continue

    elapsed = time.time() - start_time
    print(f"\nFinished loading {load_count}/{len(df)} images in {elapsed:.1f} seconds")
    print(f"Success rate: {load_count/len(df):.1%}")
    return np.array(images), np.array(labels)

X_images, y = load_images(valid_images)

# 7. Extract features and apply PCA
print("\nExtracting features with HOG + LBP and applying PCA...")
X_features = extract_enhanced_features(X_images)

# Standardize features before PCA
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_features)

# Apply PCA
pca = PCA(n_components=0.95)  # Keep 95% of variance
X_pca = pca.fit_transform(X_scaled)
print(f"Reduced features from {X_features.shape[1]} to {X_pca.shape[1]} with PCA")

# 8. Split data into 4 classes based on sample size
print("\nCreating class groups...")
def create_sample_groups(X, y, group_sizes=[25, 50, 75, 100]):
    """Split data into groups with different sample sizes"""
    unique_ids = np.unique(y)
    np.random.shuffle(unique_ids)

    groups = {}
    for size in group_sizes:
        group_ids = unique_ids[:size]
        mask = np.isin(y, group_ids)
        groups[f"{size}_classes"] = (X[mask], y[mask])

    return groups

sample_groups = create_sample_groups(X_pca, y)

# 9. Define models
models = {
    "Support Vector Machine": SVC(kernel='rbf', C=10, gamma='scale', probability=True),
    "Random Forest": RandomForestClassifier(n_estimators=100, max_depth=10, random_state=42),
    "K-Nearest Neighbors": KNeighborsClassifier(n_neighbors=5, weights='distance'),
    "Logistic Regression": LogisticRegression(multi_class='multinomial', solver='lbfgs', max_iter=1000, random_state=42),
    "Gradient Boosting": GradientBoostingClassifier(n_estimators=200, learning_rate=0.05, max_depth=3, random_state=42)
}

# 10. Train and evaluate models on each sample group
results = {}

for group_name, (X_group, y_group) in sample_groups.items():
    print(f"\n{'='*60}")
    print(f"EVALUATING ON {group_name.upper()}")
    print(f"Total samples: {len(X_group)} from {len(np.unique(y_group))} classes")

    # Split data
    X_train, X_test, y_train, y_test = train_test_split(
        X_group, y_group, test_size=0.2, random_state=42, stratify=y_group
    )

    group_results = {}

    for name, model in models.items():
        print(f"\nTraining {name}...")

        # Training
        start = time.time()
        model.fit(X_train, y_train)
        train_time = time.time() - start

        # Validation accuracy (using training data with cross-validation)
        val_accuracy = model.score(X_train, y_train)

        # Test accuracy
        start = time.time()
        y_pred = model.predict(X_test)
        pred_time = time.time() - start
        test_accuracy = accuracy_score(y_test, y_pred)

        # Store results
        group_results[name] = {
            'val_accuracy': val_accuracy,
            'test_accuracy': test_accuracy,
            'train_time': train_time,
            'pred_time': pred_time
        }

        print(f"Validation Accuracy: {val_accuracy:.4f} | Test Accuracy: {test_accuracy:.4f}")
        print(f"Training Time: {train_time:.2f}s | Prediction Time: {pred_time:.2f}s")

    results[group_name] = group_results

# 11. Display comparative results
print("\n\n" + "="*80)
print("MODEL COMPARISON SUMMARY ACROSS SAMPLE SIZES")
print("="*80 + "\n")

# Create a results table
comparison_data = []
for model_name in models.keys():
    model_row = [model_name]
    for group_name in sample_groups.keys():
        val_acc = results[group_name][model_name]['val_accuracy']
        test_acc = results[group_name][model_name]['test_accuracy']
        model_row.append(f"{val_acc:.4f}/{test_acc:.4f}")
    comparison_data.append(model_row)

# Print comparison table
headers = ["Model"] + [f"{group} (Val/Test)" for group in sample_groups.keys()]
print("{:<25} {:<20} {:<20} {:<20} {:<20}".format(*headers))
print("-"*105)
for row in comparison_data:
    print("{:<25} {:<20} {:<20} {:<20} {:<20}".format(*row))

# 12. Save the best performing model
# Find best model based on 100_classes performance
best_model_name = max(
    models.keys(),
    key=lambda x: results['100_classes'][x]['test_accuracy']
)
best_accuracy = results['100_classes'][best_model_name]['test_accuracy']

print(f"\nSaving best model ({best_model_name} with accuracy {best_accuracy:.4f} on 100 classes)...")
joblib.dump(models[best_model_name], 'best_iris_model.pkl')
print("Model saved to 'best_iris_model.pkl'")

# 13. Plot accuracy trends
plt.figure(figsize=(12, 6))
x = [25, 50, 75, 100]

for model_name in models.keys():
    test_accuracies = [results[f"{size}_classes"][model_name]['test_accuracy'] for size in x]
    plt.plot(x, test_accuracies, marker='o', label=model_name)

plt.xlabel('Number of Classes')
plt.ylabel('Test Accuracy')
plt.title('Model Performance Across Different Class Sizes')
plt.legend()
plt.grid(True)
plt.show()
```
